```python
import os
def deleteJpg(inrootpath):
    if os.path.exists(inrootpath) != True:
        print('输入的文件夹目录有误，请检查')
        return False
    try:
        for root, dirs, files in os.walk(inrootpath):
            for file in files:
                file_path = os.path.join(root, file)
                #判断后缀是不是JPG结尾，是就删除
                if str(file_path.split('.')[-1]).upper()  == 'JPG':
                    os.remove(file_path)
                    print('删除{0}照片成功'.format(file_path))
    except Exception as e:
        pass

#E:\my photo\xxx这个就是要删除的文件夹路径，把自己的文件夹路径拷贝替换上去就好了
inrootpath = r'/content/drive/MyDrive'
deleteJpg(inrootpath)
```

```python
# 文件分类
import os
import shutil

base_filename = r'/content/drive/MyDrive/人工智能/exp5/'
path = os.path.join(base_filename, 'dataset')
dst1 = os.path.join(base_filename, 'jpg2')
dst2 = os.path.join(base_filename, 'txt2')
# split_file(path, dst1, dst2)

jpg = []
txt = []
count=0
for f in os.listdir(path):
    count=count+1
    if f.endswith('.jpg'):
        jpg.append(f)
    elif f.endswith('.txt'):
        txt.append(f)

# 创建目标文件夹
if not os.path.isdir(dst1):
    os.mkdir(dst1)
if not os.path.isdir(dst2):
    os.mkdir(dst2)

# 拷贝文件到目标文件夹
for t in txt:
    _txt = os.path.join(path, t)
    shutil.copy(_txt, dst2)
for j in jpg:
    _jpg = os.path.join(path, j)
    shutil.copy(_jpg, dst1)
```

```python
# 将图片更改尺寸后保存到另一目录
from PIL import Image
import os.path
import os
filename = '/content/drive/MyDrive/人工智能/exp5/jpg2/'
def convertjpg(file,outdir,width=512,height=512):
    img=Image.open(file)
    # print(img.size)
    try:
        new_img=img.resize((width,height),Image.BILINEAR)   
        # print(new_img.size)
        new_img.save(os.path.join(outdir,os.path.basename(file)))
    except Exception as e:
        print(e)
count=0
for file in os.listdir(filename):
    # print(file)
    path=filename+file
    count=count+1
    convertjpg(path,'/content/drive/MyDrive/人工智能/exp5/jpg3')
```

```python
train_contents = []
train_guid = []
test_contents = []
test_guid = []
train_label = []
# 导入文件
import os
import io
from itertools import islice

f = open('/content/drive/MyDrive/人工智能/exp5/train.txt','r')    #读取文件
tag_dict = {}
for line in islice(f, 1, None):
    l  =  line.split(',')
    tag_dict[l[0]] = l[1].rstrip('\n')    #如果1个key有多个value，可以写成new_dict[f2[0]]  =  f2[1:]
print(len(tag_dict))
print(tag_dict.keys())  
f.close()

f = open('/content/drive/MyDrive/人工智能/exp5/test_without_label.txt','r')    #读取文件
test_dict = {}
for line in islice(f, 1, None): # 跳过第一行
    l  =  line.split(',')
    test_dict[l[0]] = l[1].rstrip('\n')
print(test_dict)
print(test_dict.keys()) 
f.close()
```

```python
start = os.listdir(r'/content/drive/MyDrive/人工智能/exp5/txt')
# start.sort(key=lambda x:int(x.split('.')[0])) #排序
print(start)
count=0
for item in start: 
    count=count+1
    train_path = '/content/drive/MyDrive/人工智能/exp5/txt/' + item
    # train_path = '/content/drive/MyDrive/信息检索/E4/train/' + item + '/'
    Result=[]
    with open(train_path, encoding='gb18030', errors='ignore') as f:
        if (item.rsplit('.txt')[0] in tag_dict.keys()):
            while True:
              line = f.readline()
              if(len(line) == 0):
                  break
              L = line.split()
              for word in L:
                  Result.append(word)
            train_contents.append(Result)
            train_guid.append(item.rsplit('.txt')[0])
            train_label.append(tag_dict[item.rsplit('.txt')[0]])

        if (item.rsplit('.txt')[0] in test_dict.keys()):
            while True:
              line = f.readline()
              if (len(line) == 0):
                  break
              L = line.split()
              for word in L:
                  Result.append(word)
            test_contents.append(Result)
            test_guid.append(item.rsplit('.txt')[0])
train=[]
for j in range(len(train_contents)):
  train.append(' '.join(train_contents[j]))
test=[]
for j in range(len(test_contents)):
  test.append(' '.join(test_contents[j]))

train_tag=[]
for i in range(len(train)):
  train_tag.append(tag_dict[train_guid[i]])
print(train_tag)

# 停用词
import random
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

stop_words = [line.strip() for line in io.open('/content/drive/MyDrive/人工智能/exp5/stopword.txt',encoding='utf-8').readlines()]
words=['ain', 'al', 'couldn', 'didn', 'doesn', 'don', 'hadn', 'hasn', 'haven', 'isn', 'll', 'mon', 'shouldn', 've', 'wasn', 'weren', 'won', 'wouldn']
for word in words:
  stop_words.append(word)

tf = TfidfVectorizer(stop_words=stop_words, max_df=1.0)
train_features = tf.fit_transform(train)
clf = MultinomialNB(alpha=1).fit(train_features, train_tag)
test_tf = TfidfVectorizer(stop_words=stop_words, max_df=1.0, vocabulary=tf.vocabulary_)
test_features = test_tf.fit_transform(test)
predicted_labels = clf.predict(test_features)
print(predicted_labels.shape)
# print(predicted_labels)
```

```python
# 验证算法准确性
rand_state = random.randrange(0,10)
x_train, x_test, y_train, y_test = train_test_split(train,train_label,test_size=0.2,random_state=rand_state)
tf = TfidfVectorizer(stop_words=stop_words, max_df=1.0)
train_features = tf.fit_transform(x_train)
test_tf = TfidfVectorizer(stop_words=stop_words, max_df=1.0, vocabulary=tf.vocabulary_)
test_features = test_tf.fit_transform(x_test)
# test_tf = TfidfVectorizer(stop_words=stop_words, max_df=1.0)
# test_features = test_tf.transform(x_test)
# 初始化朴素贝叶斯模型
clf = MultinomialNB(alpha=0.001).fit(train_features, y_train)
# 对测试集合进行预测 保存预测结果
y_predict = clf.predict(test_features)
accuracy = clf.score(test_features, y_test)
print("准确率:", accuracy)
target_names=["negative","neutral","positive"]
print(classification_report(y_test, y_predict, target_names=target_names))
```
